Transform each element of a list-like to a row, replicating index values.

DataFrame.explode(column) - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.explode.html

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'A': [[0, 1, 2], 'okay', [], [3, 4]],
                   'B': 1,
                   'C': [['a', 'b', 'c'], np.nan, [], ['d', 'e']]})
df
```

```python
df.explode('A')
```

```python
df.explode(list('AC'))
```

Explodes items like: lists, tuples, series, and np.ndarrays. The result dtype of the subset rows will be object. Empty list-likes will result in a np.nan for that row.

df.explode('column').reset_index(drop=True) - changes index to consecutive numbers
